```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_source = pd.read_csv("../file_source.csv")
```

```python
file_source.Source.value_counts()
```

```python
emodb_df = file_source[file_source.Source=="EMODB"]
```

```python
pd.Series(emodb_df.Emotions.unique()).apply(lambda x : x.title())
```

```python
plt.figure(figsize=(10,3))
# plt.title('Count of Emotions', size=16)

ax = sns.countplot(emodb_df, x="Emotions")
ax.bar_label(ax.containers[0])


plt.ylabel('Count', size=12)
plt.xlabel('Emotions', size=12)
sns.despine(top=True, right=True, left=False, bottom=False)
plt.show()
```

```python
emodb_df["Audio Length"].describe().reset_index()
```

```python
plt.figure(figsize=(10,3))
# plt.title('Audio Length', size=16)
emodb_df["Audio Length"].plot(kind="hist")
# plt.ylabel('Count', size=12)
plt.xlabel('Audio Length (seconds)', size=12)
# sns.despine(top=True, right=True, left=False, bottom=False)
plt.show()
```

```python
emodb_df["Trimmed Audio Length"].describe().reset_index()
```
